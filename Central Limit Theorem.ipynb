```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.randint(10,100)
```

```python
pop_data = [np.random.randint(10,100) for i in range(10000)]
```

```python
pop_table = pd.DataFrame({"Pop_data":pop_data})
pop_table
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Pop_data",data=pop_table)
plt.show()
```

```python
# what is the 10% of data in population
(len(pop_data)*10)/100
```

```python
sample_mean =[]
for no_sample in range(60):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
```

```python
sample_M = pd.DataFrame({"Sample_Mean":sample_mean})
```

```python
# sample_M
```

```python
sns.kdeplot(x="Sample_Mean",data=sample_M)
plt.show()
```

```python
np.mean(pop_data)
```

```python
np.mean(sample_mean)
```

The means of the samples [np.mean(sample_mean)] will be equal to the population mean.[np.mean(pop_data)]
